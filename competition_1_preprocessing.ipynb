# Reading  Data from Github

```python
import pandas as pd
import io
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
import requests
import pandas as pd

def get_github_file_contents(repo_owner, repo_name, folder_path, file_name):
    raw_url = f'https://raw.githubusercontent.com/{repo_owner}/{repo_name}/master/{folder_path}/{file_name}'
    response = requests.get(raw_url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Error: Unable to fetch data")
        return None
```

```python
repo_owner = 'aimedvedeva'
repo_name = 'Competition1'
folder_path = 'data'
train = pd.read_csv(io.StringIO(get_github_file_contents(repo_owner, repo_name, folder_path, 'train.csv')))
test = pd.read_csv(io.StringIO(get_github_file_contents(repo_owner, repo_name, folder_path, 'test.csv')))
meta_data = pd.read_csv(io.StringIO(get_github_file_contents(repo_owner, repo_name, folder_path, 'greeks.csv')))
```

# EDA

```python
def EDA(df):
    print(f'data shape: {df.shape}')
    res = pd.DataFrame(df.dtypes, columns=['data type'])
    res['#missing'] = df.isnull().sum().values
    res['%missing'] = df.isnull().sum().values / len(df)* 100
    res['#unique'] = df.nunique().values
    desc = pd.DataFrame(df.describe(include='all').transpose())
    res['min'] = desc['min'].values
    res['max'] = desc['max'].values
    res['first quartile'] = desc.loc[:, '25%'].values
    res['second quartile'] = desc.loc[:, '50%'].values
    res['third quartile'] = desc.loc[:, '75%'].values
    print(f"total missing data : {sum(res['#missing'])}")

    return res
```

```python
EDA(train)
```
